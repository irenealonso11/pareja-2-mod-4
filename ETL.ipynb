```python
# Función para extraer el modus operandi
def extraer_modus_operandi(texto):
    # Lista de palabras clave del MO
    patrones = [
        r"suffocated", r"strangling", r"shooter", r"arsonist", r"stabbe\w*", r"arsenic", r"stalked", r"kidnapped", r"abducted", r"shot\w*", r"drowned", r"Sexually assaulted"
        r"mutilat\w*", r"kill\w*", r"murder\w*", r"rap\w*", r"torture\w*", r"cannibal\w*", r"poison\w*", r"strangl\w*", r"beat", r"bite", r"dismemeberment", r"satanic cult"
    ]
    # Buscar las palabras clave en el texto y acumular las coincidencias
    modus = []
    for p in patrones:
        match = re.search(p, texto, re.IGNORECASE)
        if match:
            modus.append(match.group())  # Agregar solo la coincidencia encontrada
    # Retornar el MO encontrado, unido por comas
    return ", ".join(modus) if modus else "Unknown"
#
```

```python
# Crear la nueva columna 'modus_operandi'
df['modus_operandi'] = df['Notes'].apply(extraer_modus_operandi)

# Mostrar resultados
print(df)
```
