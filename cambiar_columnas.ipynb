```python
import pandas as pd
pd.set_option('display.max_colwidth', None)
```

```python
df = pd.read_csv("pareja-2-mod-4/serialkiller_limpio.csv",index_col=0)
```

```python
df.head()
```

```python
# Listas de palabras clave
delitos_principales = ["violación", "secuestro", "tortura", "canibalismo", "paliza", "acoso", "pirómano", "culto satánico"]
formas_matar = ["apuñalamiento", "estrangulamiento", "disparo", "veneno", "asesinato"]
# Función para extraer el delito principal
def extraer_delito_principal(texto):
    texto = texto.lower()
    for delito in delitos_principales:
        if delito in texto:
            return delito.capitalize()
    return "Otros"
# Función para extraer la forma de matar
def extraer_forma_matar(texto):
    texto = texto.lower()
    for forma in formas_matar:
        if forma in texto:
            return forma.capitalize()
    return "Asesinato"
# Crear las dos columnas nuevas aplicando las funciones
df["Delito principal"] = df["Modus_operandi"].apply(extraer_delito_principal)
df["Forma de matar"] = df["Modus_operandi"].apply(extraer_forma_matar)
# Mostrar el DataFrame resultante
print(df)
```

```python
df = df.drop(columns=["Modus_operandi"])
```

```python
# Diccionario de categorías
categorias_victimas = { 
    "menores": ["niños", "chicos jovenes", "adolescentes"],
    "mujeres": ["madre", "chicas jovenes"],
    "hombres": ["hombres", "padre", "hombres jovenes", "hombres mayores"],
    "prostitutas": ["prostitutas"],
    "homosexuales": ["homosexuales"],
    "personas_vulnerables": ["pacientes", "familias", "sin hogar", "víctimas", "personas"],
    "desconocidos": ["desconocido", "own familias"]
}

# Función para clasificar las víctimas
def clasificar_tipo_victima(texto): 
    texto = str(texto).strip().lower()
    for categoria, palabras_clave in categorias_victimas.items(): 
        for palabra in palabras_clave: 
            if palabra in texto:  # Busca coincidencias parciales
                return categoria  # Retorna la primera categoría encontrada
    return "no clasificado"  # Si no se encuentra ninguna coincidencia
    
```

```python
df["Víctimas"] = df["Tipo Víctima"].apply(clasificar_tipo_victima) # Mostrar el resultado 

print(df[["Tipo Víctima", "Víctimas"]])
```
